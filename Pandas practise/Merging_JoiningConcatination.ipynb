```python
import pandas as pd
import numpy as np
```

```python
#merging 2 data frames
```

```python
employees = pd.DataFrame({
    'emp_id':[1,2,3,4,5],
    'name':['John','Anna','Peter','Linda','Bob'],
    'department':['HR','IT','Finance','IT','HR']
}) 

#dataFrame2:salary info
salaries = pd.DataFrame({
    'emp_id':[1,2,3,6,7],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,17000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='emp_id',how='inner')
```

```python
pd.merge(employees,salaries,on='emp_id',how='outer')
```

```python
pd.merge(employees,salaries,on='emp_id',how='left')
```

```python
pd.merge(employees,salaries,on='emp_id',how='right')
```
