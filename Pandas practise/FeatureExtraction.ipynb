```python
def extract_episodes(txt):
    check =False
    data = ""
    for i in txt:
        if i ==')':
            check = False
            return data
        if check == True:
            data = data+i
        if i=='(':
            check = True
       
        
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
df
```

```python
df['Episodes']=df['Episodes'].str.replace(' eps','')
```

```python
df['Episodes']=df['Episodes'].astype(int)
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
       if txt[i] ==')':
            for j in range(i+1,i+20):
               data += txt[j]
            return data
        
```

```python
df['Total Time']=df['Title'].apply(extraction_time)
```

```python
df.head()
```

```python
def calculate_total_months(time_range):
    """
    Calculate total months from a string like 'Apr 2009 - Jul 2010'
    """
    start_str, end_str = time_range.split(' - ')
    
    start_date = pd.to_datetime(start_str, format='%b %Y')
    end_date = pd.to_datetime(end_str, format='%b %Y')
    
    total_months = (
        (end_date.year - start_date.year) * 12 +
        (end_date.month - start_date.month)
    )
    
    return total_months+1
df['Total_Months'] = df['Total Time'].apply(calculate_total_months)
```
